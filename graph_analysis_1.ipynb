```python
!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# ----- Your Dijkstra Function -----
def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0
    visited = set()
    parent = {node: None for node in graph}

    while len(visited) < len(graph):
        curr = min(
            (node for node in graph if node not in visited),
            key=lambda x: dist[x]
        )
        visited.add(curr)

        for neigh, w in graph[curr]:
            if dist[curr] + w < dist[neigh]:
                dist[neigh] = dist[curr] + w
                parent[neigh] = curr

    return dist, parent

def get_path(parent, end):
    path = []
    while end:
        path.append(end)
        end = parent[end]
    return path[::-1]

# ----- Graph -----
graph1 = {
    'A': [('B', 8), ('C', 26), ('E', 20)],
    'B': [('A', 8), ('C', 14), ('D', 17)],
    'C': [('A', 26), ('B', 14), ('D', 7), ('E', 10), ('G', 9)],
    'D': [('B', 17), ('C', 7), ('F', 5), ('G', 20)],
    'E': [('A', 20), ('C', 10), ('H', 9)],
    'F': [('D', 5), ('G', 9), ('I', 27)],
    'G': [('C', 9), ('D', 20), ('F', 9), ('H', 8), ('I', 17)],
    'H': [('E', 9), ('G', 8), ('I', 7)],
    'I': [('H', 7), ('G', 17), ('F', 27)]
}

# Run Dijkstra
dist, parent = dijkstra(graph1, 'A')
path = get_path(parent, 'I')

print("Shortest distance A to I:", dist['I'])
print("Path:", path)

# ----- Visualization -----
G = nx.Graph()

# Add weighted edges
for node in graph1:
    for neigh, w in graph1[node]:
        G.add_edge(node, neigh, weight=w)

pos = nx.spring_layout(G, seed=42)

# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=12)

# Draw edge labels (weights)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Highlight shortest path
path_edges = list(zip(path, path[1:]))

nx.draw_networkx_edges(G, pos,
                       edgelist=path_edges,
                       width=3,
                       edge_color='red')

plt.title("Dijkstra Shortest Path from A to I")
plt.show()
```
